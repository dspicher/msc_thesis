```python
groups = cPickle.load(open('groups_of_6_culture_retina_50.p','rb'))
am_sol = AmariSolve(group_size,2)
names = ['Culture','Retina']
```

```python
for key, group in groups.items():
    all_probs = [a[1] for a in group.values()]
    figure()
    for order_idx,order in enumerate([2,3,5]):
        pairs = [array([]),array([])]
        for pr in all_probs:
            me_sol = am_sol.solve(pr,order)
            pairs[0] = concatenate((pairs[0],log10(pr)))
            pairs[1] = concatenate((pairs[1],log10(me_sol)-log10(pr)))
            
        sort_idxs = argsort(pairs[0])
        log_p,error = pairs[0][sort_idxs], pairs[1][sort_idxs]
        if key==0:
            mask = log_p > -1
            scatter(log_p[mask],error[mask],c=set2[order_idx],lw=0)
        
            log_p, error = log_p[logical_not(mask)], error[logical_not(mask)]
        
        bin_size = 100
        offset=0.0
        starts = arange(0,log_p.shape[0],bin_size)
        for start in starts:
            means = (mean(log_p[start:start+bin_size]),mean(error[start:start+bin_size]))
            stds = (std(log_p[start:start+bin_size]),std(error[start:start+bin_size]))
            if start==0:
                scatter([means[0]+(order_idx-1)*offset],[means[1]],c=set2[order_idx],lw=0,alpha=1.0,label = '$P_{ME}^{('+str(order)+')}$')
            else:
                scatter([means[0]+(order_idx-1)*offset],[means[1]],c=set2[order_idx],lw=0,alpha=1.0)
            plot([means[0]+(order_idx-1)*offset-stds[0],means[0]+(order_idx-1)*offset+stds[0]],[means[1],means[1]],alpha=1.0,c=set2[order_idx])
            plot([means[0]+(order_idx-1)*offset,means[0]+(order_idx-1)*offset],[means[1]-stds[1],means[1]+stds[1]],alpha=1.0,c=set2[order_idx])
    
    if key == 1:
        leg = legend(fontsize=16,frameon=False,markerscale=1.5)
        for legobj in leg.legendHandles:
            legobj.set_alpha(1.0)
    xl,yl=xlim(),ylim()
    plot(xl,[0,0],'k--',alpha=0.5)
    xlim(xl),ylim(yl)
    xlabel('$\log_{10}(\Pr(\sigma))$',fontsize=18)
    ylabel('$\log_{10}\left(\\frac{P_{ME}(\sigma)}{\Pr(\sigma)}\\right)$',fontsize=18)
    a=yticks()
    yticks(a[0],fontsize=12)
    a=xticks()
    xticks(a[0],fontsize=12)
    title(names[key],fontsize=18)
    gca().spines['top'].set_visible(False)
    gca().spines['right'].set_visible(False)
    tight_layout()    
    gca().tick_params(axis='x',which='major',top='off')
    gca().tick_params(axis='y',which='major',right='off')
    savefig('me_scatter_{0}.pdf'.format(names[key]))
```

```python
import matplotlib.gridspec as gridspec

active_neurons=array([sum([int(a) for a in bin(nr)[2:]]) for nr in range(64)])
rand_margin = 0.3
tickfs = 16
labelfs = 20
labelmathfs = 24
for groupKey, group in groups.items():
    all_probs = [a[1] for a in group.values()]
    figure(figsize=(18.0/5*4,4))
    
    gs = gridspec.GridSpec(1, 5,width_ratios=[3,3,3,1,1])
    
    sca(subplot(gs[0]))
    act_probs_sums = [[] for _ in range(7)]
    act_probs = [array([]) for _ in range(7)]
    offset = 0.35
    spacing = 0.4
    for pr in all_probs:
        #ppl.scatter(log10(pr),-offset+rand(pr.shape[0])*rand_margin-rand_margin/2+active_neurons,color='k',alpha=0.3)
        for act in range(7):
            act_probs[act] = concatenate((act_probs[act],log10(pr[active_neurons==act])))
            act_probs_sums[act].append(np.sum(pr[active_neurons==act]))
    ppl.boxplot(act_probs,vert=False,positions=spacing*arange(7))
    for act in range(7):
        scatter(median(log10(act_probs_sums[act])),spacing*(offset+act),s=20,c='k',lw=0,marker='D')
        plot([min(log10(act_probs_sums[act])),max(log10(act_probs_sums[act]))],spacing*array([offset+act,offset+act]),c='k')
            
    xticks(range(-4,1), fontsize=tickfs)
    yticks(spacing*arange(7),[str(a) for a in range(7)],fontsize=tickfs)
    ylim([-spacing/2,6.5*spacing])
    xl = xlim()
    xlim([xl[0],0.1])
    ylabel('$\\left| \\{\\sigma_i \\vert \\sigma_i=1\\}\\right|$',fontsize=labelmathfs)
    xlabel('$\log_{10}\Pr(\sigma)$',fontsize=labelmathfs)
    gca().spines['bottom'].set_visible(True)
    gca().tick_params(axis='x',which='major',bottom='on')
    
    sca(subplot(gs[1]))
    all_etas = [array([]) for act in range(6)]
    spacing = 0.5
    for pr in all_probs:
        etas = fit_log_linear_etas(pr)
        for act in range(1,7):
            pts = array([etas[key] for key in filter(lambda k: sum(k)==act,etas.keys())])
            all_etas[act-1] = concatenate((all_etas[act-1],pts))
    ppl.boxplot(all_etas,vert=False,positions=spacing*arange(1,7))
    xlabel('$\eta_I$',fontsize=labelmathfs)
    ylabel('$\\left| I\\right|$',fontsize=labelmathfs)
    x = xticks()
    xticks(x[0][::2], fontsize=tickfs)
    yticks(spacing*arange(1,7),[str(a) for a in range(1,7)],fontsize=tickfs)
    ylim([1*spacing-spacing/2,6.5*spacing])
    xl = xlim()
    xlim([0.0,xl[1]])
    gca().spines['bottom'].set_visible(True)
    gca().tick_params(axis='x',which='major',bottom='on')
    
    sca(subplot(gs[2]))
    all_ints = [array([]) for act in range(6)]
    for pr in all_probs:
        ints = fit_log_linear_interactions(pr)
        for act in range(1,7):
            pts = array([ints[key] for key in filter(lambda k: sum(k)==act,ints.keys())])
            all_ints[act-1] = concatenate((all_ints[act-1],pts))
    ppl.boxplot(all_ints,vert=False,positions=spacing*arange(1,7))
    xlabel('$\\theta_I$',fontsize=labelmathfs)
    ylabel('$\\left| I\\right|$',fontsize=labelmathfs)
    x = xticks()
    xticks(x[0][::2], fontsize=tickfs)
    yticks(spacing*arange(1,7),[str(a) for a in range(1,7)],fontsize=tickfs)
    ylim([1*spacing-spacing/2,6.5*spacing])
    gca().spines['bottom'].set_visible(True)
    gca().tick_params(axis='x',which='major',bottom='on')
    
    sca(subplot(gs[3]))
    entropies = [entropy(pr) for pr in all_probs]
    ppl.boxplot(entropies)
    gca().tick_params(axis='y',which='major',left='on')
    xticks([])
    yt = yticks()
    yticks(yt[0][::2],fontsize=tickfs)
    ylabel('entropy [bits]',fontsize=labelfs)
    gca().spines['bottom'].set_visible(False)
    
    sca(subplot(gs[4]))
    am_sol = AmariSolve(6,2)
    multi_infs = [entropy(am_sol.solve(pr,1))-entropy(pr) for pr in all_probs]
    ppl.boxplot(multi_infs)
    gca().tick_params(axis='y',which='major',left='on')
    xticks([])
    yt = yticks()
    yticks(yt[0][::2],fontsize=tickfs)
    ylabel('multi-information [bits]',fontsize=labelfs)
    gca().spines['bottom'].set_visible(False)
    
    tight_layout()
    suptitle(names[groupKey],fontsize=25)
    subplots_adjust(right=1.02,left=0.05,bottom=0.19,top=0.95)
    savefig('dataset_characterization_{0}.pdf'.format(names[groupKey]))
```

```python
for key, group in groups.items():
    all_probs = [a[1] for a in group.values()]
    entropies = zeros((len(all_probs),group_size))
    for group_idx in range(len(all_probs)):
        for order in range(1,group_size+1):
            entropies[group_idx,order-1] = entropy(am_sol.solve(all_probs[group_idx],order))
            
    mask = array([np.all(array(list(reversed(sorted(entropies[i,:])))==entropies[i,:])) for i in range(entropies.shape[0])])
    to_pl = []
    for curr_ord in range(1,group_size-1):
        to_pl.append((entropies[mask,0]-entropies[mask,curr_ord])/(entropies[mask,0]-entropies[mask,-1]))
    figure()
    ppl.boxplot(to_pl,widths=0.3)
    title(str(key))
    ylim([0.75,1.05])
    xticks(range(1,5),[r'$P_{ME}^{('+str(i)+')}$' for i in range(2,6)],fontsize=18)
    xl = xlim()
    plot(xl,[1,1],'k--',alpha=0.4)
    xlim(xl)
    ylabel('$I_k$',fontsize=18)
    a=yticks()
    yticks(a[0],fontsize=12)
    ylim([0.75,1.02])
    title(names[key],fontsize=18)
    tight_layout()
    tick_params(axis='y',which='major',left='on')
    savefig('multi_inf_{0}.pdf'.format(names[key]))
```

```python
i = 0
am_sol = AmariSolve(6,2)
for key, group in groups.items():
    all_probs = [a[1] for a in group.values()]
    pr = all_probs[i]
    ind_ent = entropy(am_sol.solve(pr,1))
    multi_inf = ind_ent-entropy(pr)
    etas = {i:{} for i in range(7)}
    ints = {i:{} for i in range(7)}
    all_me_likelis = []
    fracs = []
    for order in range(7):
        me_sol = am_sol.solve(pr,order)
        all_me_likelis.append(me_sol)
        if 1 < order < 6:
            fracs.append((ind_ent - entropy(me_sol))/multi_inf)
        curr_etas = fit_log_linear_etas(me_sol)
        etas[order].update(curr_etas)
        curr_ints = fit_log_linear_interactions(me_sol)
        ints[order].update(curr_ints)

import brewer2mpl
set2 = brewer2mpl.get_map('Set2', 'qualitative', 8).mpl_colors
figure(figsize=(20,5))
labelfs = 24
labelmathfs = 28
tickfs = 20
subplot(1,3,1)
label_done = [False for _ in range(7)]
active_neurons=array([sum([int(a) for a in bin(nr)[2:]]) for nr in range(64)])
for act_i,act in enumerate(active_neurons):
    if not label_done[act]:
        ppl.plot(log10(array([p[act_i] for p in all_me_likelis])),'.-',color=set2[act],label=str(act))
        label_done[act] = True
    else:
        ppl.plot(log10(array([p[act_i] for p in all_me_likelis])),'.-',color=set2[act])
    
xlim([-0.5,6.5])
xticks(range(0,7,2),[r'$P_{ME}^{('+str(i)+')}$' for i in range(0,7,2)],fontsize=labelfs)
yt = yticks()
yticks(yt[0][::2],fontsize=tickfs)
ylabel('$\log_{10}(\Pr(\sigma))$',fontsize=labelmathfs)
leg = legend(loc='upper left', bbox_to_anchor=(3.2, 1.0),fontsize=18,frameon=False)
for legobj in leg.legendHandles:
    legobj.set_linewidth(3.0)
tick_params(axis='y',which='major',left='on')
gca().spines['bottom'].set_visible(False)

subplot(1,3,2)
for key in sorted(etas[1].keys()):
    vals = [etas[i][key] for i in range(7)]
    ppl.plot(vals,'.-',color=set2[sum(key)])
xlim([-0.5,6.5])
xticks(range(0,7,2),[r'$P_{ME}^{('+str(i)+')}$' for i in range(0,7,2)],fontsize=labelfs)
yt = yticks()
yticks(yt[0][::2],fontsize=tickfs)
ylabel('$\eta_I$',fontsize=labelmathfs)
for iFrac,frac in enumerate(fracs):
    text(iFrac+2,0.7,"{:.3f}".format(frac),horizontalalignment='center',fontsize=16)
text(3.5,0.77,'$I_k$',fontsize=20,horizontalalignment='center')
tick_params(axis='y',which='major',left='on')
gca().spines['bottom'].set_visible(False)

subplot(1,3,3)
for key in sorted(ints[1].keys()):
    vals = [ints[i][key] for i in range(7)]
    ppl.plot(vals,'.-',color=set2[sum(key)])
xlim([-0.5,6.5])
xticks(range(0,7,2),[r'$P_{ME}^{('+str(i)+')}$' for i in range(0,7,2)],fontsize=labelfs)
yt = yticks()
yticks(yt[0][::2],fontsize=tickfs)
ylabel('$\\theta_I$',fontsize=labelmathfs)
#legend(loc='upper left', bbox_to_anchor=(0.98, 1.0),fontsize=18)
tick_params(axis='y',which='major',left='on')
gca().spines['bottom'].set_visible(False)
#tight_layout()
subplots_adjust(wspace=0.12,left=0.045,right=0.95,top=0.95)
savefig('me_progression_params.pdf')
```
