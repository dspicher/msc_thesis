```python
def get_data(name,idx,order,hid):
    data = cPickle.load(open('rbm_fits/fin_{0}_{1}_order_{2}_hid_{3}.p'.format(name,idx,order,hid),'rb'))
    if np.sum(data['return']==0)==0:
        try:
            data = cPickle.load(open('rbm_fits/cleanup_{0}_{1}_order_{2}_hid_{3}_succ_5.p'.format(name,idx,order,hid),'rb'))
        except:
            pass
    return data
```

```python
def get_rbm_params(params,vis,hid):
    visbiases = params[:vis]
    hidbiases = params[vis:vis+hid]
    vishid = np.reshape(params[vis+hid:], (vis,hid))
    return visbiases,hidbiases,vishid


def get_probabilities(vishid, visbiases, hidbiases):
    data = create_binary_patterns(visbiases.shape[0])
    all_hid = create_binary_patterns(hidbiases.shape[0])
    visbiases_energy = np.reshape(np.sum(data*np.reshape(visbiases, (1, -1)), 1), (-1, 1))
    hidbiases_energy = np.reshape(np.sum(all_hid*np.reshape(hidbiases, (1, -1)), 1), (1, -1))
    interactions_energy = np.dot(np.dot(data, vishid), all_hid.T)
    pmass = np.exp(visbiases_energy + hidbiases_energy + interactions_energy)
    return np.sum(pmass/np.sum(pmass),1)


def create_binary_patterns(n):
    nrs = np.arange(2**n)
    return np.bool8(np.floor(pow(2.0, np.arange(-n+1, 1, 1))*np.reshape(nrs, (-1, 1))) % 2)
```

```python
names = ['Culture','Retina']
orders = [1,2,4,6]
hids = [1,2,5,10]
N = 50

def get_successful_fits(data):
    succ = data['return']==0
    return data['final_obj_func'][succ]

def get_successful_count(data):
    return np.sum(data['return']==0)

def get_best_rbm_probs(data):
    succ = data['return']==0
    if np.sum(succ)==0:
        return None
    best = argmin(data['final_obj_func'][succ])
    params = data['xopt'][succ,:]
    params = params[best,:]
    visb, hidb, vishid = get_rbm_params(params,6,data['hid'])
    return get_probabilities(vishid,visb,hidb)
```

```python
def get_conn_cumulant(pr,idxs):
    n = log2(pr.shape[0])
    patterns = create_binary_patterns(n)[:,idxs]
    centered = patterns-reshape(np.sum(reshape(pr,(-1,1))*patterns,0),(1,-1))
    return np.sum(pr*np.prod(centered,1),0)

from itertools import combinations
def get_all_conn_cumulants(pr):
    n = int(round(log2(pr.shape[0])))
    return {i:array([get_conn_cumulant(pr,idxs) for idxs in combinations(range(n),i)]) for i in range(2,n+1)}
```
